### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
date=[]
category =[]
title=[]
for d in soup.find(class_="part_list_2").find_all('h3'):
    date_=d.find(class_="date").text
    category_= d.find("em").text
    title_= d.find("a").text
    print(date_,category_,title_)
    date.append(date_)
    category.append(category_)
    title.append(title_)
    
```
